```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

spark = SparkSession.builder \
    .appName("Week5 Assignment") \
    .master("local[*]") \
    .getOrCreate()

df = spark.read.csv(
    "Sample - Superstore.csv",
    header=True,
    inferSchema=True
)

df.show(5)
```

```python
"""
Q1. What are the key limitations of traditional MapReduce that make Spark a preferred choice for modern big data processing?

Answer:
• MapReduce stores intermediate results on disk, making processing slower.
• It has high disk I/O overhead.
• It is not suitable for iterative machine learning algorithms.
• Multiple MapReduce jobs are needed for complex workflows.
• Spark performs computations in memory, making it much faster.
• Spark also supports SQL, Machine Learning, Streaming, and Graph Processing within one framework.
"""
```

```python
"""
Q2. Explain how Spark uses In-Memory Computing to speed up iterative machine learning algorithms compared to disk-based systems.

Answer:
Spark stores intermediate data in RAM instead of writing it to disk after every operation. Since machine learning algorithms repeatedly process the same data, keeping the data in memory avoids repeated disk access and significantly reduces execution time.
"""
```

```python
"""
Q3. Remove duplicate rows based on user_id and transaction_date.
"""

data = [
    (101,"2024-01-01",500),
    (101,"2024-01-01",600),
    (102,"2024-02-10",700),
    (103,"2024-03-12",800)
]

df_users = spark.createDataFrame(
    data,
    ["user_id","transaction_date","amount"]
)

print("Before Removing Duplicates")
df_users.show()

df_users = df_users.dropDuplicates(["user_id","transaction_date"])

print("After Removing Duplicates")
df_users.show()
```

```python
"""
Q4. Filter the West region and calculate the average sale amount for each product category.
"""

from pyspark.sql.functions import avg
df.filter(df.Region == "West") \
  .groupBy("Category") \
  .agg(avg("Sales").alias("Average Sales")) \
  .show()
```

```python
#Q6. Find the total count of records for each city, but only for cities where the count is greater than 100.
print("\nQ6 Output")

city_count = (
    df.groupBy("City")
      .agg(
          count("*").alias("Total_Count")
      )
      .filter(
          col("Total_Count") > 100
      )
)

city_count.show()
```

```python
# Q7. How does the immutability of Spark DataFrames affect data cleaning?

# Answer:

# Spark DataFrames are immutable, meaning they cannot be modified after creation.
# Every transformation such as dropping columns, renaming columns,
# or filtering rows creates a new DataFrame rather than modifying the original one.

new_df = df.drop("column_name")
new_df = new_df.withColumnRenamed("old_name", "new_name")
print("\nQ7 Example")

new_df = df.drop("Postal Code")
new_df = new_df.withColumnRenamed(
    "Customer Name",
    "Customer_Name"
)

new_df.show(5)
```

```python
"""
Q8. Filter rows where age is between 18 and 30 (inclusive)
and subscription is 'Premium'.
"""
print("\nQ8 Output")

demo_data = [
    ("Alice", 20, "Premium"),
    ("Bob", 25, "Basic"),
    ("Charlie", 30, "Premium"),
    ("David", 35, "Premium"),
    ("Eva", 18, "Premium")
]

demo_df = spark.createDataFrame(
    demo_data,
    ["Name", "Age", "Subscription"]
)

result = demo_df.filter(
    (col("Age").between(18, 30)) &
    (col("Subscription") == "Premium")
)

result.show()
```

```python
"""
Q9. When cleaning a dataset, why is it often better to handle null values
before performing mathematical aggregations like sum() or avg()?

Answer:

Null values can affect the accuracy of mathematical operations.
Replacing or removing null values before performing aggregations
ensures correct totals, averages, minimums, and maximums.
It also prevents unexpected errors during data processing.
"""
```

```python
"""
Q10. Cast the column raw_timestamp to TimestampType
and rename it to event_time.
"""

print("\nQ10 Output")

timestamp_data = [
    ("2024-01-10 10:20:30",),
    ("2024-05-11 12:30:00",),
    ("2024-08-25 15:45:10",)
]

timestamp_df = spark.createDataFrame(
    timestamp_data,
    ["raw_timestamp"]
)

timestamp_df = timestamp_df.withColumn(
    "event_time",
    col("raw_timestamp").cast(TimestampType())
).drop("raw_timestamp")

timestamp_df.show(truncate=False)
```

```python
"""
Q11. Explain the Shuffle process during grouping.

Answer:

A shuffle is the process where Spark redistributes data across
different partitions so that rows with the same key are brought
together.

Operations such as groupBy(), join(), distinct(), and reduceByKey()
trigger a shuffle.

It is called a wide transformation because data moves across
partitions instead of remaining within a single partition.
Shuffling is expensive because it involves network communication
and disk I/O.
"""
```

```python
"""
Q12. Remove rows where email is null OR username is empty.
"""

print("\nQ12 Output")

user_data = [
    ("abc@gmail.com", "Alice"),
    (None, "Bob"),
    ("xyz@gmail.com", ""),
    ("john@gmail.com", "John"),
    ("mary@gmail.com", "Mary")
]

user_df = spark.createDataFrame(
    user_data,
    ["email", "username"]
)

print("Before Cleaning")
user_df.show()

clean_df = user_df.filter(
    col("email").isNotNull() &
    (trim(col("username")) != "")
)

print("After Cleaning")
clean_df.show()
```

```python
"""
Q13. Calculate multiple statistics (Minimum, Maximum and Average)
using the agg() function.
"""
from pyspark.sql.functions import min, max, mean

df.agg(
    min("Sales").alias("Minimum Sales"),
    max("Sales").alias("Maximum Sales"),
    mean("Sales").alias("Average Sales")
).show()
```

```python
"""
Q14. What is the risk of using inferSchema=True when the dataset
contains inconsistent date formats?

Answer:

When date formats are inconsistent, Spark may infer the wrong
data type or treat the entire column as a String.

This can lead to:
• Incorrect date parsing
• Null values after conversion
• Errors during filtering and aggregation
• Incorrect analytical results

In production systems, it is generally better to define the schema
explicitly or convert date columns using functions such as
to_date() or to_timestamp().
"""
```

```python
"""
Q15. Build a processing pipeline that:

1. Removes duplicate rows.
2. Fills null Sales values with 0.
3. Groups by Region.
4. Calculates total revenue.
"""

print("\nQ15 Output")

pipeline = (
    df.dropDuplicates()
      .na.fill({"Sales": 0})
      .groupBy("Region")
      .agg(
          sum("Sales").alias("Total_Revenue")
      )
      .orderBy(desc("Total_Revenue"))
)
pipeline.show()

print("\nOverall Dataset Statistics")

df.select(
    count("*").alias("Total_Rows"),
    sum("Sales").alias("Total_Sales"),
    avg("Sales").alias("Average_Sales"),
    min("Sales").alias("Minimum_Sales"),
    max("Sales").alias("Maximum_Sales")
).show()
spark.stop()
```
